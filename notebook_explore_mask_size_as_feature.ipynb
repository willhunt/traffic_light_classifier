# Explore mask size as a feature to improve traffic light color classifiaction

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

## Check probability function to apply
# Normal probability density function

```python
mean = 0
scale = 1
x = np.linspace(-5, 5)
y = norm.pdf(x, mean, scale)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, y)
fig.suptitle("PDF with loc = {0}, scale = {1}".format(mean, scale))
```

## Scale for use with 25x25 pixel images

```python
mean = 50 / (25*25)
scale = 0.1
x = np.linspace(0, 1)
y = norm.pdf(x, mean, scale)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, y)
fig.suptitle("PDF with loc = {0}, scale = {1}".format(mean, scale))
```

### Sigmoid function
Alternative function

```python
def sigmoid(X, x_scale=1, x_shift=0):
    X_new = (X + x_shift) * x_scale 
    return 1 / (1 + np.exp(-X_new))
```

```python
x = np.linspace(0, 1)
x_shift = 0.4
x_scale = -25
y = sigmoid(x, x_scale, x_shift)
fig, ax = plt.subplots(1, 1)
ax.plot(x, y)
fig.suptitle("Sigmoid with x-scale = {0}, x-shift = {1}".format(x_scale, x_shift))
```
